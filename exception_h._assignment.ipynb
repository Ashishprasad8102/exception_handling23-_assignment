```python
1. Python, an exception is an event that occurs during the execution of a program that disrupts the normal
flow of the program's instructions. When an exception occurs, it typically results in the termination of
the program, unless it is handled properly. 

 Syntax errors are detected by the Python interpreter during the parsing phase, before the program is executed.
Exceptions, on the other hand, occur during the execution of the program when certain
conditions are met.
```

```python
2.When an exception is not handled in Python, it results in the termination of the program and an error message
is displayed, indicating the type of exception that occurred and the traceback information. This abrupt 
termination without proper handling can leave the program in an inconsistent state and prevent it from 
completing its intended tasks.
#without exeception

def divide_numbers(a, b):
    result = a / b
    return result
num1 = 10
num2 = 0

# Call the function without exception handling
result = divide_numbers(num1, num2)
print(result)



#with exception
def divide_numbers(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")

num1 = 10
num2 = 0

# Call the function with exception handling
result = divide_numbers(num1, num2)
print(result)

```

```python
3.#try & else
def divide_numbers(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")

num1 = 10
num2 = 0

# Call the function with exception handling
result = divide_numbers(num1, num2)
print(result)

In Python, the try and else statements are used together to handle exceptions in a more controlled and 
structured manner. The try block allows you to enclose a section of code that might raise an exception, 
while the else block allows you to specify code that should be executed if no exception occurs within
the try block.


#finaly
In Python, the finally block is used in conjunction with the try and except blocks to define code that 
should be executed regardless of whether an exception occurred or not. The finally block ensures that 
certain cleanup operations or essential actions are performed, whether an exception is raised or
not.

def divide_numbers(a, b):
    try:
        result = a / b
        print("The division result is:", result)
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
    finally:
        print("Cleanup: Division operation completed.")

# Test cases
divide_numbers(10, 2)  # No exception occurs
divide_numbers(10, 0)  # Raises ZeroDivisionError



#raise
In Python, the raise statement is used to explicitly raise an exception. It allows you to create and
raise custom exceptions or propagate built-in exceptions in specific situations. The raise statemen
t is followed by an exception object or class that represents the type of exception being raised.



class CustomException(Exception):
    pass

def process_data(data):
    if not data:
        raise CustomException("No data available!")
    else:
        # Process the data

# Test case
data = []
try:
    process_data(data)  # Raises CustomException
except CustomException as e:
    print("An exception occurred:", str(e))
    
```

```python
5.Custom exceptions in Python are user-defined exception classes that inherit from the base Exception
class or one of its subclasses. They allow developers to create specific types of exceptions that are
tailored to their application's requirements and provide meaningful error messages or additional
attributes. 

class InvalidInputException(Exception):
    def __init__(self, message):
        self.message = message

def process_data(data):
    if not data:
        raise InvalidInputException("Invalid input data!")
    else:
        # Process the data

# Test case
data = []
try:
    process_data(data)  # Raises InvalidInputException
except InvalidInputException as e:
    print("An exception occurred:", str(e))
```
